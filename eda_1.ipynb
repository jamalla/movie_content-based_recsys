```python
import ast
```

```python
def convert(obj):
    L= []
    
    for i in ast.literal_eval(obj):
        L.append(i["name"])
        
    return L
```

```python
movies_df_2["genres"] = movies_df_2["genres"].apply(convert)
movies_df_2["keywords"] = movies_df_2["keywords"].apply(convert)
```

```python
movies_df_2.head()
```

```python
def convert3(obj):
    L= []
    counter = 0
    
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i["name"])
            counter+=1
        else:
            break
        
    return L
```

```python
movies_df_2["cast"] = movies_df_2["cast"].apply(convert3)
```

```python
movies_df_2.head()
```

```python
def fetch_director(obj):
    L= []
    
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            
    return L
```

```python
movies_df_2["crew"] = movies_df_2["crew"].apply(fetch_director)
```

```python
movies_df_2["overview"] = movies_df_2["overview"].apply(lambda x:x.split())
```

```python
movies_df_2["genres"] = movies_df_2["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies_df_2["keywords"] = movies_df_2["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies_df_2["cast"] = movies_df_2["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies_df_2["crew"] = movies_df_2["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(movies_df_final["tags"]).toarray().shape
```

```python
vectors = cv.fit_transform(movies_df_final["tags"]).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem())
    
    return y
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])), reverse= True, key= lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = movies_df_final[movies_df_final["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse= True, key= lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies_df_final.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
